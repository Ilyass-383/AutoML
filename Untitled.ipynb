```python
# ================================
# 1. IMPORTS
# ================================
import csv
import itertools
from collections import Counter

from river import compose, preprocessing, tree, drift, metrics

# ================================
# 2. STREAM SIMULATOR
# ================================
def data_stream(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)

        for row in reader:
            # cible
            y = int(row.pop('label'))

            # retirer colonnes qui causent une fuite d'information
            row.pop('attack_cat', None)
            row.pop('id', None)

            # conversion automatique des colonnes numériques
            for k, v in row.items():
                try:
                    row[k] = float(v)
                except:
                    pass

            yield row, y

# ================================
# 3. VERIFICATION DES CLASSES
# ================================
counter = Counter()

for _, y in itertools.islice(data_stream('UNSW_NB15_training_shuffled.csv'), 5000):
    counter[y] += 1

print("Distribution des classes sur les 5000 premières lignes :")
print(counter)

# ================================
# 4. MODELE
# ================================
model = compose.Pipeline(
    preprocessing.OneHotEncoder(),
    tree.HoeffdingTreeClassifier(
        grace_period=200,
        delta=1e-7,
        leaf_prediction='mc',
        nb_threshold=100
    )
)

# ================================
# 5. DRIFT DETECTOR + METRICS
# ================================
adwin = drift.ADWIN()

metric_acc = metrics.Accuracy()
metric_f1 = metrics.F1()
metric_precision = metrics.Precision()
metric_recall = metrics.Recall()
metric_kappa = metrics.CohenKappa()

drift_count = 0

# ================================
# 6. STREAM
# ================================
gen = data_stream('UNSW_NB15_training_shuffled.csv')

# ================================
# 7. PHASE 1 : APPRENTISSAGE NORMAL
# ================================
print("\n=== PHASE 1 : APPRENTISSAGE NORMAL ===")

for i, (x, y) in enumerate(itertools.islice(gen, 5000)):

    # prédiction
    pred = model.predict_one(x)

    # mise à jour métriques
    if pred is not None:
        metric_acc.update(y, pred)
        metric_f1.update(y, pred)
        metric_precision.update(y, pred)
        metric_recall.update(y, pred)
        metric_kappa.update(y, pred)

    # apprentissage
    model.learn_one(x, y)

    # affichage tous les 500 exemples
    if i % 500 == 0:
        print(
            f"[Phase 1] Instance {i} | "
            f"Acc={metric_acc.get():.4f} | "
            f"F1={metric_f1.get():.4f} | "
            f"Precision={metric_precision.get():.4f} | "
            f"Recall={metric_recall.get():.4f} | "
            f"Kappa={metric_kappa.get():.4f}"
        )

# ================================
# 8. PHASE 2 : DRIFT INJECTE
# ================================
print("\n=== 🚨 PHASE 2 : DÉRIVE INJECTÉE ===")

for i, (x, y) in enumerate(itertools.islice(gen, 2000)):

    # dérive artificielle : inversion des labels
    y_drifted = 1 - y

    # prédiction
    pred = model.predict_one(x)

    # mise à jour métriques
    if pred is not None:
        metric_acc.update(y_drifted, pred)
        metric_f1.update(y_drifted, pred)
        metric_precision.update(y_drifted, pred)
        metric_recall.update(y_drifted, pred)
        metric_kappa.update(y_drifted, pred)

        # erreur pour ADWIN
        error = 0 if pred == y_drifted else 1
        adwin.update(error)

        # détection de drift
        if adwin.drift_detected:
            drift_count += 1

            print(f"\n⚠️ Drift détecté à l'instance {i}")

            # reset modèle
            model = compose.Pipeline(
                preprocessing.OneHotEncoder(),
                tree.HoeffdingTreeClassifier(
                    grace_period=200,
                    delta=1e-7,
                    leaf_prediction='mc',
                    nb_threshold=100
                )
            )

            # reset ADWIN
            adwin = drift.ADWIN()

    # apprentissage sur les labels driftés
    model.learn_one(x, y_drifted)

    # affichage tous les 250 exemples
    if i % 250 == 0:
        print(
            f"[Phase 2] Instance {i} | "
            f"Acc={metric_acc.get():.4f} | "
            f"F1={metric_f1.get():.4f} | "
            f"Precision={metric_precision.get():.4f} | "
            f"Recall={metric_recall.get():.4f} | "
            f"Kappa={metric_kappa.get():.4f}"
        )

# ================================
# 9. RESULTATS FINAUX
# ================================
print("\n=== RESULTATS FINAUX ===")
print(f"Accuracy finale            : {metric_acc.get():.4f}")
print(f"F1 final                   : {metric_f1.get():.4f}")
print(f"Precision finale           : {metric_precision.get():.4f}")
print(f"Recall final               : {metric_recall.get():.4f}")
print(f"Kappa final                : {metric_kappa.get():.4f}")
print(f"Nombre de drifts détectés  : {drift_count}")
```

```python
import numpy as np
from sklearn.linear_model import SGDClassifier

class SGDWrapper:
    def __init__(self):
        self.model = SGDClassifier(
            loss='log_loss',
            learning_rate='adaptive',
            eta0=0.01,
            random_state=42
        )

        self.feature_names = None
        self.category_maps = {}
        self.is_fitted = False

    def _encode_value(self, feature_name, value):
        # Si numérique
        if isinstance(value, (int, float)):
            return float(value)

        # Si catégorielle
        value = str(value)

        if feature_name not in self.category_maps:
            self.category_maps[feature_name] = {}

        if value not in self.category_maps[feature_name]:
            self.category_maps[feature_name][value] = len(self.category_maps[feature_name])

        return float(self.category_maps[feature_name][value])

    def _to_array(self, x: dict) -> np.ndarray:
        if self.feature_names is None:
            self.feature_names = list(x.keys())

        values = []

        for f in self.feature_names:
            v = x.get(f, 0.0)
            values.append(self._encode_value(f, v))

        return np.array(values, dtype=float).reshape(1, -1)

    def learn_one(self, x: dict, y: int):
        X = self._to_array(x)

        if not self.is_fitted:
            self.model.partial_fit(X, [y], classes=[0, 1])
            self.is_fitted = True
        else:
            self.model.partial_fit(X, [y])

    def predict_one(self, x: dict) -> int:
        if not self.is_fitted:
            return 0

        X = self._to_array(x)
        return int(self.model.predict(X)[0])
```

```python
# ================================
# 1. IMPORTS
# ================================
import json
import time
import itertools
import numpy as np

from sklearn.linear_model import SGDClassifier
from river import tree, neighbors, metrics, compose, preprocessing

from src.stream_simulator import data_stream


# ================================
# 2. SGD WRAPPER
# ================================
class SGDWrapper:
    def __init__(self):
        self.model = SGDClassifier(
            loss='log_loss',
            learning_rate='adaptive',
            eta0=0.01,
            random_state=42
        )

        self.feature_names = None
        self.category_maps = {}
        self.is_fitted = False

    def _encode(self, x):
        if self.feature_names is None:
            self.feature_names = list(x.keys())

        values = []

        for f in self.feature_names:
            v = x.get(f, 0.0)

            if isinstance(v, (int, float)):
                values.append(float(v))
            else:
                v = str(v)

                if f not in self.category_maps:
                    self.category_maps[f] = {}

                if v not in self.category_maps[f]:
                    self.category_maps[f][v] = len(self.category_maps[f])

                values.append(float(self.category_maps[f][v]))

        return np.array(values, dtype=float).reshape(1, -1)

    def learn_one(self, x, y):
        X = self._encode(x)

        if not self.is_fitted:
            self.model.partial_fit(X, [y], classes=[0, 1])
            self.is_fitted = True
        else:
            self.model.partial_fit(X, [y])

    def predict_one(self, x):
        if not self.is_fitted:
            return 0

        X = self._encode(x)
        return int(self.model.predict(X)[0])


# ================================
# 3. MODELES
# ================================

# Hoeffding Tree
model_ht = compose.Pipeline(
    preprocessing.OneHotEncoder(),
    tree.HoeffdingTreeClassifier(
        grace_period=100,
        delta=1e-5,
        leaf_prediction='nba',
        nb_threshold=30
    )
)

# KNN plus léger
model_knn = compose.Pipeline(
    preprocessing.OneHotEncoder(),
    neighbors.KNNClassifier(
        n_neighbors=3,
        engine=neighbors.SWINN(
            maxlen=100
        )
    )
)

# SGD
model_sgd = SGDWrapper()

models = {
    "HT": model_ht,
    "KNN": model_knn,
    "SGD": model_sgd
}


# ================================
# 4. METRIQUES
# ================================
results = {
    name: {
        "acc": [],
        "kappa": [],
        "time_ms": []
    }
    for name in models
}

metrics_dict = {
    name: metrics.Accuracy() + metrics.CohenKappa()
    for name in models
}


# ================================
# 5. STREAMING EVALUATION
# ================================
MAX_SAMPLES = 3000

stream = data_stream('UNSW_NB15_training_shuffled.csv')

for i, (x, y) in enumerate(itertools.islice(stream, MAX_SAMPLES)):

    for name, model in models.items():

        # temps de prédiction
        t0 = time.perf_counter()
        pred = model.predict_one(x)
        latency = (time.perf_counter() - t0) * 1000  # ms

        # mise à jour métriques
        if pred is not None:
            metrics_dict[name].update(y, pred)

        # apprentissage
        model.learn_one(x, y)

        # sauvegarde des stats tous les 100 exemples
        if i % 100 == 0:
            results[name]["acc"].append(metrics_dict[name][0].get())
            results[name]["kappa"].append(metrics_dict[name][1].get())
            results[name]["time_ms"].append(latency)

    # affichage progression
    if i % 500 == 0:
        print(f"\n===== Instance {i} =====")

        for name in models:
            print(
                f"{name} | "
                f"Acc={metrics_dict[name][0].get():.4f} | "
                f"Kappa={metrics_dict[name][1].get():.4f}"
            )


# ================================
# 6. RESULTATS FINAUX
# ================================
print("\n===== RESULTATS FINAUX =====")

for name in models:
    acc = metrics_dict[name][0].get()
    kappa = metrics_dict[name][1].get()
    avg_time = np.mean(results[name]["time_ms"])

    print(
        f"{name} | "
        f"Accuracy={acc:.4f} | "
        f"Kappa={kappa:.4f} | "
        f"Temps moyen={avg_time:.4f} ms"
    )


# ================================
# 7. SAUVEGARDE JSON
# ================================
with open("baseline_eval.json", "w", encoding="utf-8") as f:
    json.dump(results, f, indent=4)

print("\n✅ Evaluation terminée")
print("📁 Résultats sauvegardés dans baseline_eval.json")
```

```python
# Test jouet du bandit avant intégration
import random, math

class UCB1Bandit:
    def __init__(self, n_arms):
        self.n = n_arms
        self.counts = [0] * n_arms    # nb fois chaque bras tiré
        self.values = [0.0] * n_arms  # reward moyen de chaque bras
        self.total = 0

    def select(self) -> int:
        # Phase d'initialisation : essaye chaque bras une fois
        for i in range(self.n):
            if self.counts[i] == 0:
                return i
        # UCB1 : score + bonus d'exploration
        ucb = [self.values[i] +
               math.sqrt(2 * math.log(self.total) / self.counts[i])
               for i in range(self.n)]
        return ucb.index(max(ucb))

    def update(self, arm: int, reward: float):
        self.total += 1
        self.counts[arm] += 1
        # Moyenne mobile incrémentale
        n = self.counts[arm]
        self.values[arm] += (reward - self.values[arm]) / n
```

```python
# src/auto_ml.py

import logging
from river import metrics
from src.bandit import UCB1Bandit


class AutoMLSystem:
    def __init__(self, models, model_names, update_interval=100):
        self.models = models
        self.names = model_names
        self.bandit = UCB1Bandit(n_arms=len(models))
        self.metric = metrics.Accuracy()

        self.current_arm = 0
        self.current_pred_arm = 0
        self.switch_log = []
        self.n_instances = 0
        self.update_interval = update_interval

        logging.basicConfig(
            filename='automl_switches.log',
            level=logging.INFO,
            format='%(asctime)s - %(message)s'
        )

    def predict_and_learn(self, x, y):

        # Re-sélection du modèle seulement toutes les N instances
        if self.n_instances % self.update_interval == 0:
            arm = self.bandit.select()
            self.current_pred_arm = arm

            if arm != self.current_arm:
                switch_info = {
                    'instance': self.n_instances,
                    'from': self.names[self.current_arm],
                    'to': self.names[arm]
                }

                self.switch_log.append(switch_info)

                logging.info(
                    f"Switch à l'instance {self.n_instances}: "
                    f"{self.names[self.current_arm]} -> {self.names[arm]}"
                )

                self.current_arm = arm

        arm = self.current_pred_arm

        # prédiction avec le modèle sélectionné
        pred = self.models[arm].predict_one(x)

        if pred is None:
            pred = 0

        reward = int(pred == y)

        # mise à jour bandit
        self.bandit.update(arm, reward)

        # tous les modèles apprennent
        for model in self.models:
            model.learn_one(x, y)

        # update métrique globale
        self.metric.update(y, pred)

        self.n_instances += 1

        return pred
```
